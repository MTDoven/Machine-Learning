```python
from Library.UtilsTools.DataGenerater import  get_classify_data
from Library.Criterion import LogarithmicLoss
from Library.Module import SequentialModule
from Library.Activator import Sigmoid
from Library.BPLayer import Linear
```

```python
X,y = get_classify_data(size=(100,4), classes=2)
```

```python
class Model(SequentialModule):
    def __init__(self, loss):
        self.Structure = [Linear(4,4),
                          Linear(4,1),
                          Sigmoid()]
        self.Lr = 0.00001
        self.Loss = loss
```

```python
criterian = LogarithmicLoss()
model = Model(criterian)
```

```python
for _ in range(10000):
    result = model(X)
    loss = criterian(result,y)
    model.backward()
```
