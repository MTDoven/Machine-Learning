# Kmean iris

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from kmeans import KMeans
```

```python
iris = load_iris()
X_iris = iris.data
y_true = iris.target
scaler = StandardScaler()
X_iris = scaler.fit_transform(X_iris)
```

```python
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X_iris)
silhouette_avg = silhouette_score(X_iris, labels)
print(f'Silhouette Score: {silhouette_avg:.2f}')
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_iris)

# Create a figure with 1 row and 2 columns for the subplots
fig, ax = plt.subplots(1, 2, figsize=(20, 7))

# Plot KMeans clustering results
for i in range(3):
    ax[0].scatter(X_pca[labels == i, 0], X_pca[labels == i, 1], label=f'Cluster {i}')
ax[0].set_xlabel('PCA Component 1')
ax[0].set_ylabel('PCA Component 2')
ax[0].set_title('KMeans Clustering of Iris Data')
ax[0].legend()

# Plot true labels
for i in range(3):
    ax[1].scatter(X_pca[y_true == i, 0], X_pca[y_true == i, 1], label=iris.target_names[i])
ax[1].set_xlabel('PCA Component 1')
ax[1].set_ylabel('PCA Component 2')
ax[1].set_title('True Labels of Iris Data')
ax[1].legend()

plt.show()
```

# Kmeans MNIST

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from kmeans import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load MNIST dataset
mnist = fetch_openml('mnist_784', version=1)
X_mnist = mnist.data
y_true = mnist.target.astype(int)  # Convert target to integers

# Standardize the data
scaler = StandardScaler()
X_mnist = scaler.fit_transform(X_mnist)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=10)
labels = kmeans.fit_predict(X_mnist)
silhouette_avg = silhouette_score(X_mnist, labels)
print(f'Silhouette Score: {silhouette_avg:.2f}')

# t-SNE 
pca = PCA(n_components=2) 
X_pca = pca.fit_transform(X_mnist)

# Create a figure with 1 row and 2 columns for the subplots
fig, ax = plt.subplots(1, 2, figsize=(20, 7))

# Plot KMeans clustering results
for i in range(10):
    ax[0].scatter(X_pca[labels == i, 0], X_pca[labels == i, 1], label=f'Cluster {i}')
ax[0].set_xlabel('PCA Component 1')
ax[0].set_ylabel('PCA Component 2')
ax[0].set_title('KMeans Clustering of MNIST Data')
ax[0].legend()

# Plot true labels
for i in range(10):
    ax[1].scatter(X_pca[y_true == i, 0], X_pca[y_true == i, 1], label=f'Digit {i}')
ax[1].set_xlabel('PCA Component 1')
ax[1].set_ylabel('PCA Component 2')
ax[1].set_title('True Labels of MNIST Data')
ax[1].legend()

plt.show()
```

# FCM MNIST

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from fcmeans import FCM

mnist = fetch_openml('mnist_784')
X_mnist = mnist.data
y_true = mnist.target
scaler = StandardScaler()
X_mnist = scaler.fit_transform(X_mnist)

# Apply FCM clustering
fcm = FCM(n_clusters=10)  # MNIST has 10 classes
fcm.fit(X_mnist)
labels = fcm.predict(X_mnist)
fcm_silhouette_avg = silhouette_score(X_mnist, labels)
print(f'FCM Silhouette Score: {fcm_silhouette_avg:.2f}')

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_mnist)

# Create a figure with 1 row and 2 columns for the subplots
fig, ax = plt.subplots(1, 2, figsize=(20, 7))

# Plot FCM clustering results
for i in range(10):
    ax[0].scatter(X_pca[labels == i, 0], X_pca[labels == i, 1], label=f'Cluster {i}')
ax[0].set_xlabel('PCA Component 1')
ax[0].set_ylabel('PCA Component 2')
ax[0].set_title('FCM Clustering of MNIST Data')
ax[0].legend()

# Plot true labels
for i in range(10):
    ax[1].scatter(X_pca[y_true == i, 0], X_pca[y_true == i, 1], label=f'Digit {i}')
ax[1].set_xlabel('PCA Component 1')
ax[1].set_ylabel('PCA Component 2')
ax[1].set_title('True Labels of MNIST Data')
ax[1].legend()

plt.show()
```
