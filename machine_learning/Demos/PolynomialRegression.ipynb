```python
from Library.UtilsTools.DataGenerater import get_linear_data
X,y = get_linear_data(size=(1000,2), seed=5)
y = (y/100)**4 - (y/100)**3 - 5*(y/100)**2
```

```python
from Library.Algorithms import PolynomialRegression
polynomial_regression = PolynomialRegression(inn=2, out=1, power=6)
polynomial_regression.Structure[1].RElevel=0.001
```

```python
loss = polynomial_regression.train(X,y,Lr=0.000001,epoch=50000)
loss
```

```python
y_pred = polynomial_regression.evaluate(X)
y_pred
```
